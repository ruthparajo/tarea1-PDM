# Limpieza texto

```python
import re
```

```python
import spacy
nlp = spacy.load("es_core_news_sm")

def eliminar_stopwords(texto):
    doc = nlp(texto)
    palabras_limpias = [token.text for token in doc if not token.is_stop and not token.is_punct]
    return " ".join(palabras_limpias)
```

```python
def lematizar_texto(texto):
    doc = nlp(texto)
    lemas = [token.lemma_ for token in doc if not token.is_punct]
    return " ".join(lemas)
```

```python
def limpiar_texto(txt):
    txt = re.sub(r"@\w+", "", txt)
    txt = re.sub(r"[\n,\t]", "", txt)
    txt = re.sub(r"\\n", "", txt)
    txt = txt.lower()
    txt = re.sub(r'[\u0300-\u036f]', '', txt)
    txt = eliminar_stopwords(txt)
    #txt = lematizar_texto(txt)
    return txt.strip()
```

# Extraer keywords

```python
from keybert import KeyBERT

kw_model = KeyBERT()

def get_keywords(text, top_n=3):
    # Devuelve lista de (keyword, score)
    keywords = kw_model.extract_keywords(text, top_n=top_n)
    return [k[0] for k in keywords]
```

### Test

```python
test_df["keywords"] = test_df["limpio"].apply(lambda x: get_keywords(x, top_n=3))
test_df[["limpio", "keywords"]].head()
```

```python
import unicodedata

def quitar_acentos(texto):
    texto = unicodedata.normalize('NFD', texto)
    texto = re.sub(r'[\u0300-\u036f]', '', texto)
    return texto

def quitar_acentos_lista(lista_texto):
    return [quitar_acentos(t) for t in lista_texto]
```

```python
def filtrar_keywords_base(keywords):
    resultado = []
    for palabra in keywords:
        if not any(palabra != otra and palabra in otra for otra in keywords):
            resultado.append(palabra)
    return resultado
keywords = ['fiscalización', 'indicacion', 'indicaciones']
print(filtrar_keywords_base(keywords))
```

```python
test_df['keywords'] = test_df['keywords'].apply(lambda x: quitar_acentos_lista(x))
test_df['keywords'] = test_df['keywords'].apply(lambda x: filtrar_keywords_base(x))
```

### Para toda la base

```python
bases = {'part_desc_2023_1': part_desc_2023_1, 'part_desc_2023_2': part_desc_2023_2, 'part_desc_2024': part_desc_2024}
```

```python
for k,base in bases.items():
    print('Haciendo', k)
    base['texto_limpio'] = base['TEXTO_PRINCIPAL'].apply(limpiar_texto)
    print('Limpieza de texto hecha', k)
    base["keywords"] = base["texto_limpio"].apply(lambda x: get_keywords(x, top_n=3))
    base['keywords'] = base['keywords'].apply(lambda x: quitar_acentos_lista(x))
    base['keywords'] = base['keywords'].apply(lambda x: filtrar_keywords_base(x))
    print('Keywords extraidos', k)
    base.to_csv(f'{k}_keywords.csv', index=False)
    
```
